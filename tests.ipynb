# Testing and Babbling

```python
import eyelib
import numpy as np
import matplotlib.pyplot as plt
```

```python
coefs = (2, 3, 4)
a = np.poly1d(coefs)
x = np.linspace(-2, 2)

plt.plot(x, a(x))
plt.show()
```

```python
coefs_prime = eyelib.move_second_degree(coefs, 1, 1)
coefs_prime
```

```python
b = np.poly1d(coefs_prime)

plt.plot(x, a(x))
plt.plot(x, b(x))
plt.show()
```
